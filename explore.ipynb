```python
import duckdb

query = ''' 
    select
    puzzle_number,
    answer,
    count(*) as times_used
    from raw_wordle_words
    group by puzzle_number ,answer
    having count(*) > 1'''
with duckdb.connect("./data/wordle.db") as con:
    display(con.query(query).df())
    
```
